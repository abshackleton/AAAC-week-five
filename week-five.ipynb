```python
info = info[info['Rank'] < 11]
info.head()
```

```python
info.mean(numeric_only=True)
```

```python
info.groupby('Continent')[["Average size (feet)"]].agg(['count', 'min', 'max'])
```

```python
info = info.drop(columns=['Geographical range', 'Conservation status'])
info.head()
```

```python
info['Average size (m)'] = info['Average size (feet)'].map(lambda x: x / 3.281)
info.head()
```

```python
info['Size range (m)'] = info.apply(lambda x: x['Longest length (feet)'] - x['Minimum size (feet)'] / 3.281, axis=1)
info.head()
```

```python
info.reset_index(drop=True, inplace=True)
info.head()
```
